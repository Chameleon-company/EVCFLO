```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('New_Australian_EV_Charging_Stations.csv')
```

```python
df
```

```python
df.describe()
```

- "Charging station" and "Nearby EVStations" have wide distribution.
- "plugs_Type2", "Plugs_Three_Phase", "Plugs-CHAdeMO" are having low values, and some outliners.
- "Hospitals", "Parks", "Restaurant", "Malls","Supermarkets" are having high means, shows that many locations have these amenities.

## Check different types of plugs related to different facility

```python
##drop the columns
df_dropped = df.drop(['Unnamed: 0', 'Latitude', 'Longitude'], axis=1)

##Pearson's Correlation
pearson_corr = df_dropped.corr(method='pearson')
print("Pearson's correlation matrix:")
print(pearson_corr)

```

```python
##Spearman's Correlation
spearman_corr = df_dropped.corr(method='spearman')
print("\nSpearman's correlation matrix:")
print(spearman_corr)
```

## 1:Plus_Type2

```python
numeric_columns = [
    'Plugs_Type2','Power 1', 'charging_stations',
    'Nearby EVStations', 'Hospitals', 'Parks', 'Restaurants', 'Malls', 'Supermarkets'
]


# Calculate the correlation matrix
correlation_matrix = df[numeric_columns].corr()

# Plot the heatmap
plt.figure(figsize=(12, 9))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
