# Format GeoJson

Extracts the required coordinates required to scan bounding box's with the Google Maps Automatic scan. GeoJson file containnig bounding boxs's can be generated using (geojson.io), saved as a GeoJson file and then extracted here.

```python
pip install geojson
```

```python
import numpy as np
import pandas as pd
import geojson
from datetime import date
```

```python
#Provide the name of the geojson file with bounding boxs to be scanned marked out
file_name = "aus_nz.geojson"
today = date.today()
#If source of data being contained in bounding box's is something else replace Google_API
source = "Google_API"
source_date = today.strftime("%d/%m/%Y")
```

```python
#load the file and prep array
with open(file_name) as f:
    gj = geojson.load(f)
features = gj['features']
column_names = ['west', 'east', 'north', 'south', 'source', 'source_date']
corners = []

for i in features:
    
    north = i['geometry']['coordinates'][0][2][1]
    south = i['geometry']['coordinates'][0][0][1]
    east = i['geometry']['coordinates'][0][1][0]
    west = i['geometry']['coordinates'][0][0][0]

    corners.append([west, east, north, south, source, source_date])
    
df = pd.DataFrame(corners, columns = column_names)
df
```

```python
# Save to coordinates file to be used for geojson scane
df.to_csv("Files/Coordinates2.csv")
```
