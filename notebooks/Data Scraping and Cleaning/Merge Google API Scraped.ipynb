# Merge Google API Scraped
#### Use to merge all of the files generated when scraping them using Google API (EV Chargers) - GeoJson. Also appended to the end of each row is the current date and source (Google API). As they are helpful in determining the reliability of the data on the website.
#### Finally also add source and source date to all columns Goggle API and current date

```python
import pandas as pd
import numpy as np
from datetime import date
from datetime import timedelta
```

```python
#Specify the number of files containing locations
count = 35

# Replace location with place you wish to read from
location = 'Files/'
# Replace name with whatever name you used
name = 'test_name'

df = pd.DataFrame()
all_df = []
```

```python
for i in range(1, count):
    df = pd.read_csv(location+name+str(i-1)+".csv")
    all_df.append(df)

df = pd.concat(all_df)
```

```python
df = df.drop("Unnamed: 0",axis=1)
df = df.drop_duplicates()
df = df.reset_index(drop=True)
```

```python
df.to_csv(name+"_Merged.csv", index=False)
```
